```python
import numpy as np
from deflenclass import LMS_def_len
import lineecho
```

```python
h = lineecho.m1()

N= 64

lms = LMS_def_len(realizations=100, 
          iterations=50000, 
          sigma_x=1, 
          alpha=0.5,
          sigma_r=1E-6, 
          step_size=2/(30*N), 
          system_response=h, 
          initial_solution=np.zeros(N),
          decimation_factor=1,
          noise_type='gaussian')
```

```python
Jinftheo, Vhisttheo, Jhisttheo, msdhisttheo = lms.predict()
```

```python
Vhistsim, e2histsim, msdhistsim = lms.run()
```

```python
e2vartheo = lms.steadystateerrorvariance()
lms.plot_mse(mse_simulation=e2histsim, mse_theoretical=Jhisttheo, mse_steady_state=Jinftheo, mse_variance_theo=e2vartheo)
```

```python
lms.plot_msd(msd_simulation=msdhistsim, msd_theoretical=msdhisttheo)
```
